4. **function**: APIì˜ ì´ì „ ë²„ì „ì—ì„œ ì‚¬ìš©ëë˜ ê²ƒìœ¼ë¡œ, íŠ¹ì • ê¸°ëŠ¥ì´ë‚˜ ë™ì‘ì„ ì–´ì‹œìŠ¤í„´íŠ¸ì—ê²Œ ìš”ì²­í•  ë•Œ ì‚¬ìš©ë˜ì—ˆìŠµë‹ˆë‹¤. ìµœì‹  ë²„ì „ì—ì„œëŠ” ì¼ë°˜ì ìœ¼ë¡œ ì‚¬ìš©ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤.

ì´ëŸ¬í•œ ì—­í• ì€ ëŒ€í™”í˜• ì„¸ì…˜ì—ì„œ ë©”ì‹œì§€ì˜ ë°œì‹ ìì™€ ì˜ë„ë¥¼ êµ¬ë³„í•˜ëŠ” ë° ë„ì›€ì„ ì¤ë‹ˆë‹¤. APIëŠ” ì´ ì •ë³´ë¥¼ ì‚¬ìš©í•˜ì—¬ ì ì ˆí•œ ë°©ì‹ìœ¼ë¡œ ë°˜ì‘í•˜ê±°ë‚˜ ì‘ë‹µí•©ë‹ˆë‹¤.

```python
completion = openai.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello!"}
  ]
)

print(completion.choices[0].message)
```

```python
# assistant ì— ëŒ€í•œ ì§€ì¹¨ì´ ë˜ëŠ” context ì •ë³´ë¥¼ ì œê³µ
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
]

while True:
    # ì‚¬ìš©ìë¡œë¶€í„° ë©”ì‹œì§€ ì…ë ¥ ë°›ê¸°
    message = input("User : ")

    # ì‚¬ìš©ìê°€ 'stop'ì´ë¼ê³  ì…ë ¥í•˜ë©´ ëŒ€í™” ì¢…ë£Œ
    if message.lower() == "stop":
        break
    else:
        # ì‚¬ìš©ìì˜ ë©”ì‹œì§€ë¥¼ ë©”ì‹œì§€ ëª©ë¡ì— ì¶”ê°€
        messages.append(
             {"role": "user", "content": message}
        )

        # OpenAI APIë¥¼ ì‚¬ìš©í•˜ì—¬ ëŒ€í™”í˜• ì™„ì„± ì§„í–‰
        completion = openai.chat.completions.create(
          model="gpt-3.5-turbo",
          messages=messages
        )

    # ëª¨ë¸ì˜ ë‹µë³€ ì¶”ì¶œ
    reply = completion.choices[0].message.content

    # ëª¨ë¸ì˜ ë‹µë³€ ì¶œë ¥
    print(f"ChatGPT: {reply}")

    # ëª¨ë¸ì˜ ë‹µë³€ì„ ë©”ì‹œì§€ ëª©ë¡ì— ì¶”ê°€
    messages.append({"role": "assistant", "content": reply})
```

BearerëŠ” HTTP ì¸ì¦ ìŠ¤í‚¤ë§ˆ ì¤‘ í•˜ë‚˜ì…ë‹ˆë‹¤. ì›¹ ì„œë¹„ìŠ¤ì—ì„œ ì‚¬ìš©í•˜ëŠ” í† í° ê¸°ë°˜ì˜ ì¸ì¦ ë°©ì‹ ì¤‘ ê°€ì¥ ì¼ë°˜ì ì¸ ë°©ì‹ì…ë‹ˆë‹¤.
```python
import requests  

# OpenAI APIì˜ URLì„ ì •ì˜
URL = "https://api.openai.com/v1/chat/completions"

# ìš”ì²­ì— í•„ìš”í•œ ë°ì´í„°ë¥¼ payload ë³€ìˆ˜ì— ì €ì¥
payload = {
    "model": "gpt-3.5-turbo",  
    "messages": [{"role": "user", "content": f"Who is the most successful chess champion in the world ?"}],  
    "temperature": 1.0,  
    "top_p": 1.0,  
    "n": 1,  
    "stream": False,     # ìŠ¤íŠ¸ë¦¼ëª¨ë“œ ì‚¬ìš© ì—¬ë¶€
    "presence_penalty": 0,  # ê²°ê³¼ì˜ ì¼ê´€ì„±ì— ì˜í–¥
    "frequency_penalty": 0,  # ë‹¨ì–´ì˜ ì‚¬ìš© ë¹ˆë„ì— ì˜í–¥
}

# ìš”ì²­ í—¤ë”ì— ë‚´ìš© ìœ í˜• ë° ì¸ì¦ í‚¤ ì„¤ì •
headers = {
    "Content-Type": "application/json",  # ìš”ì²­ ë³¸ë¬¸ì˜ ìœ í˜•ì„ JSONìœ¼ë¡œ ì§€ì •
    "Authorization": f"Bearer {openai.api_key}"  # ì¸ì¦ í‚¤ í¬í•¨
}

# requests.postë¥¼ ì‚¬ìš©í•˜ì—¬ OpenAI APIì— POST ìš”ì²­ì„ ë³´ëƒ…ë‹ˆë‹¤.
response = requests.post(URL, headers=headers, json=payload, stream=False)
```
