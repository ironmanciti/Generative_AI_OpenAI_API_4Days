# 컨볼루셔널 변이형 오토인코더

이 노트북은 MNIST 데이터세트에서 변이형 오토인코더(VAE, Variational Autoencoder)를 훈련하는 방법을 보여줍니다. VAE는 오토인코더의 확률론적 형태로, 높은 차원의 입력 데이터를 더 작은 표현으로 압축하는 모델입니다. 입력을 잠재 벡터에 매핑하는 기존의 오토인코더와 달리 VAE는 입력 데이터를 가우스 평균 및 분산과 같은 확률 분포의 매개변수에 매핑합니다. 이 방식은 연속적이고 구조화된 잠재 공간을 생성하므로 이미지 생성에 유용합니다.

```python
from IPython import display

import glob
import imageio
import matplotlib.pyplot as plt
import numpy as np
import PIL
import tensorflow as tf
import time
```

## MNIST 데이터세트 로드하기

각 MNIST 이미지는 원래 각각 0-255 사이인 784개의 정수로 구성된 벡터이며 픽셀 강도를 나타냅니다. 모델에서 Bernoulli 분포를 사용하여 각 픽셀을 모델링하고 데이터세트를 정적으로 이진화합니다.

```python
(train_images, _), (test_images, _) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
def preprocess_images(images):
    # 이미지 배열의 모양을 변경하고 255로 나누어 정규화합니다.
    images = images.reshape((images.shape[0], 28, 28, 1)) / 255.
    # 이미지 픽셀 값이 0.5보다 크면 1.0으로, 아니면 0.0으로 변환합니다.
    return np.where(images > .5, 1.0, 0.0).astype('float32')

# 학습 이미지 데이터 전처리
train_images = preprocess_images(train_images)
# 테스트 이미지 데이터 전처리
test_images = preprocess_images(test_images)
```

```python
train_size = 60000
batch_size = 32
test_size = 10000
```

## *tf.data*를 사용하여 데이터 배치 및 셔플 처리하기

```python
train_dataset = (tf.data.Dataset.from_tensor_slices(train_images)
                 .shuffle(train_size).batch(batch_size))
test_dataset = (tf.data.Dataset.from_tensor_slices(test_images)
                .shuffle(test_size).batch(batch_size))
```

## *tf.keras.Sequential*을 사용하여 인코더 및 디코더 네트워크 정의하기

이 VAE 예제에서는 인코더 및 디코더 네트워크에 두 개의 작은 ConvNet을 사용합니다. 문헌에서, 이들 네트워크는 각각 추론/인식 및 생성 모델로도 지칭됩니다. 구현을 단순화하기 위해 `tf.keras.Sequential`을 사용합니다. 다음 설명에서 $x$ 및 $z$는 각각 관측 값과 잠재 변수를 나타냅니다.

### 인코더 네트워크

이것은 근사 사후 분포 $q(z|x)$를 정의합니다. 이 분포는 관측 값을 입력으로 받고 잠재 표현 $z$의 조건부 분포를 지정하기 위한 매개변수 세트를 출력합니다. 이 예에서는 분포를 대각선 가우스로 간단히 모델링하고 네트워크는 인수 분해된 가우스의 평균 및 로그-분산 매개변수를 출력합니다. 수치 안정성을 위해 분산을 직접 출력하지 않고 로그-분산을 출력합니다.

### 디코더 네트워크

잠재 샘플 $z$를 입력으로 사용하여 관측 값의 조건부 분포에 대한 매개변수를 출력하는 관측 값 $p(x|z)$의 조건부 분포를 정의합니다. 잠재 이전 분포 $p(z)$를 단위 가우스로 모델링합니다.

### 재매개변수화 트릭

훈련 중에 디코더에 대해 샘플 $z$를 생성하기 위해, 입력 관측 값 $x$가 주어졌을 때 인코더에 의해 출력된 매개변수로 정의된 잠재 분포로부터 샘플링할 수 있습니다. 그러나 역전파가 무작위 노드를 통해 흐를 수 없기 때문에 이 샘플링 작업에서 병목 현상이 발생합니다.

이를 해결하기 위해 재매개변수화 트릭을 사용합니다. 이 예에서는 디코더 매개변수와 다른 매개변수 $\epsilon$을 다음과 같이 사용하여 $z$를 근사시킵니다.

$$z = \mu + \sigma \odot \epsilon$$

여기서 $\mu$ 및 $\sigma$는 각각 가우스 분포의 평균 및 표준 편차를 나타냅니다. 이들은 디코더 출력에서 파생될 수 있습니다. $\epsilon$은 $z$의 무질서도를 유지하는 데 사용되는 무작위 노이즈로 생각할 수 있습니다. 표준 정규 분포에서 $\epsilon$을 생성합니다.  
$\odot$: 요소별 곱셈(element-wise multiplication)

잠재 변수 $z$는 이제 $\mu$, $\sigma$ 및 $\epsilon$의 함수에 의해 생성되며, 이를 통해 모델은 각각 $\mu$ 및 $\sigma$를 통해 인코더의 그래디언트를 역전파하면서 $\epsilon$을 통해 무질서도를 유지할 수 있습니다.

### 네트워크 아키텍처

인코더 네트워크의 경우 두 개의 컨볼루션 레이어, 그리고 이어서 완전히 연결된 레이어를 사용합니다. 디코더 네트워크에서 완전히 연결된 레이어와 그 뒤에 세 개의 컨볼루션 전치 레이어(일부 컨텍스트에서는 디컨볼루션 레이어라고도 함)를 사용하여 이 아키텍처를 미러링합니다. 미니 배치 사용으로 인한 추가 무질서도가 샘플링의 무질서에 더해 불안정성을 높일 수 있으므로 VAE 훈련시 배치 정규화를 사용하지 않는 것이 일반적입니다.


```python
class CVAE(tf.keras.Model):
  """Convolutional variational autoencoder."""

  def __init__(self, latent_dim):
    super(CVAE, self).__init__()
    self.latent_dim = latent_dim
    # 인코더 정의
    self.encoder = tf.keras.Sequential(
        [
            tf.keras.layers.InputLayer(input_shape=(28, 28, 1)),
            tf.keras.layers.Conv2D(
                filters=32, kernel_size=3, strides=(2, 2), activation='relu'),
            tf.keras.layers.Conv2D(
                filters=64, kernel_size=3, strides=(2, 2), activation='relu'),
            tf.keras.layers.Flatten(),
            # No activation
            tf.keras.layers.Dense(latent_dim + latent_dim),
        ]
    )

    # 디코더 정의
    self.decoder = tf.keras.Sequential(
        [
            tf.keras.layers.InputLayer(input_shape=(latent_dim,)),
            tf.keras.layers.Dense(units=7*7*32, activation=tf.nn.relu),
            tf.keras.layers.Reshape(target_shape=(7, 7, 32)),
            tf.keras.layers.Conv2DTranspose(
                filters=64, kernel_size=3, strides=2, padding='same',
                activation='relu'),
            tf.keras.layers.Conv2DTranspose(
                filters=32, kernel_size=3, strides=2, padding='same',
                activation='relu'),
            # No activation
            tf.keras.layers.Conv2DTranspose(
                filters=1, kernel_size=3, strides=1, padding='same'),
        ]
    )

  @tf.function
  def sample(self, eps=None):
    # 잠재 공간에서 샘플링된 벡터 eps를 받아 디코딩
    # eps가 제공되지 않으면 잠재 공간의 크기에 따라 무작위 벡터를 생성합니다.
    if eps is None:
      eps = tf.random.normal(shape=(100, self.latent_dim))
    return self.decode(eps, apply_sigmoid=True)

  def encode(self, x):
    # 입력 x를 인코딩하여 평균과 로그 분산을 반환
    mean, logvar = tf.split(self.encoder(x), num_or_size_splits=2, axis=1)
    return mean, logvar

  def reparameterize(self, mean, logvar):
    # 재매개변수화 트릭을 사용하여 샘플 z를 생성
    # 평균과 로그 분산을 기반으로 정규 분포에서 무작위 샘플을 취한 후
    # 이를 변형하여 z를 생성
    eps = tf.random.normal(shape=mean.shape)
    return eps * tf.exp(logvar * .5) + mean

  def decode(self, z, apply_sigmoid=False):
    # 잠재 변수 z를 입력으로 받아 디코딩하여 이미지 로짓을 반환
    logits = self.decoder(z)
    # apply_sigmoid 플래그가 True로 설정된 경우, 로짓에 시그모이드를 적용하여 확률을 반환
    if apply_sigmoid:
      probs = tf.sigmoid(logits)
      return probs
    return logits
```

## 손실 함수 및 옵티마이저 정의하기

VAE는 한계 로그-우도에 대한 ELBO(evidence lower bound)를 최대화하여 훈련합니다.

$$\log p(x) \ge \text{ELBO} = \mathbb{E}_{q(z|x)}\left[\log \frac{p(x, z)}{q(z|x)}\right].$$

실제로, 이 예상에 대한 단일 샘플 Monte Carlo 추정값을 최적화합니다.

$$\log p(x| z) + \log p(z) - \log q(z|x),$$ 여기서 $z$는 $q(z|x)$에서 샘플링됩니다.

참고: KL 항을 분석적으로 계산할 수도 있지만 여기서는 단순화를 위해 Monte Carlo 예측 도구에 세 항을 모두 통합합니다.

```python
optimizer = tf.keras.optimizers.Adam(1e-4)

def log_normal_pdf(sample, mean, logvar, raxis=1):
    log2pi = tf.math.log(2. * np.pi)  # 2π의 로그값을 계산
    # 주어진 평균 및 로그 분산에 대한 샘플의 로그 정규 분포 확률 밀도를 계산
    return tf.reduce_sum(
        -.5 * ((sample - mean) ** 2. * tf.exp(-logvar) + logvar + log2pi),
        axis=raxis)

def compute_loss(model, x):
    # x를 인코딩하여 평균과 로그 분산을 얻습니다.
    mean, logvar = model.encode(x)
    # 재매개변수화 트릭을 사용하여 z를 샘플링합니다.
    z = model.reparameterize(mean, logvar)
    # z를 디코딩하여 복원된 이미지 (로짓)를 얻습니다.
    x_logit = model.decode(z)
    # 원본 이미지 x와 복원된 이미지 x_logit 간의 크로스 엔트로피를 계산합니다.
    cross_ent = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_logit, labels=x)
    logpx_z = -tf.reduce_sum(cross_ent, axis=[1, 2, 3])
    logpz = log_normal_pdf(z, 0., 0.)
    logqz_x = log_normal_pdf(z, mean, logvar)
    # VAE의 손실 함수를 계산합니다.
    return -tf.reduce_mean(logpx_z + logpz - logqz_x)

@tf.function
def train_step(model, x, optimizer):
    with tf.GradientTape() as tape:
        loss = compute_loss(model, x)
    # 손실에 대한 그래디언트를 계산
    gradients = tape.gradient(loss, model.trainable_variables)
    # 그래디언트를 사용하여 모델의 가중치를 업데이트
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
```

## 훈련하기

- 데이터세트를 반복하여 시작합니다.
- 반복하는 동안 매번 이미지를 인코더로 전달하여 근사적인 사후 $q(z|x)$의 평균 및 로그-분산 매개변수 세트를 얻습니다.
- 그런 다음 $q(z|x)$에서 샘플링하기 위해 *재매개변수화 트릭*을 적용합니다.
- 마지막으로, 생성된 분포 $p(x|z)$의 로짓을 얻기 위해 재매개변수화된 샘플을 디코더로 전달합니다.
- 참고: 훈련 세트에 60k 데이터 포인트와 테스트 세트에 10k 데이터 포인트가 있는 keras에 의해 로드된 데이터세트를 사용하기 때문에 테스트세트에 대한 결과 ELBO는 Larochelle MNIST의 동적 이진화를 사용하는 문헌에서 보고된 결과보다 약간 높습니다.

### 이미지 생성하기

- 훈련을 마쳤으면 이미지를 생성할 차례입니다.
- 우선, 단위 가우스 사전 분포 $p(z)$에서 잠재 벡터 세트를 샘플링합니다.
- 그러면 생성기가 잠재 샘플 $z$를 관측 값의 로짓으로 변환하여 분포 $p(x|z)$를 제공합니다.
- 여기서 Bernoulli 분포의 확률을 플롯합니다.


```python
epochs = 10
latent_dim =  2  #시각화하려면 잠재 공간의 차원성을 평면(2차원)으로 설정
# 생성할 이미지의 예제 수를 16으로 설정
num_examples_to_generate = 16

# 생성(예측)을 위해 무작위 벡터를 상수로 유지합니다. 따라서 개선 사항을 쉽게 볼 수 있습니다.
random_vector_for_generation = tf.random.normal(
    shape=[num_examples_to_generate, latent_dim])
# 지정된 잠재 차원을 사용하여 CVAE 모델을 초기화
model = CVAE(latent_dim)
```

```python
def generate_and_save_images(model, epoch, test_sample):
  # test_sample을 인코딩하여 평균 및 로그 분산을 얻습니다.
  mean, logvar = model.encode(test_sample)
  # 재매개변수화 트릭을 사용하여 z를 샘플링
  z = model.reparameterize(mean, logvar)
  # z를 사용하여 이미지를 생성
  predictions = model.sample(z)

  fig = plt.figure(figsize=(4, 4))
  for i in range(predictions.shape[0]):
    plt.subplot(4, 4, i + 1)
    plt.imshow(predictions[i, :, :, 0], cmap='gray')
    plt.axis('off')

  plt.tight_layout()
  # 이미지를 파일로 저장합니다. 파일명은 현재 에포크 번호를 포함합니다.
  plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
  plt.show()
```

```python
# 출력 이미지를 생성하기 위해 테스트 세트에서 샘플을 선택
# 배치 크기가 생성하려는 예제 수보다 크거나 같아야 함을 확인
assert batch_size >= num_examples_to_generate

# 테스트 데이터셋에서 한 배치만 가져옵니다.
for test_batch in test_dataset.take(1):
  # 해당 배치에서 원하는 수만큼의 예제를 선택하여 test_sample에 저장
  test_sample = test_batch[0:num_examples_to_generate, :, :, :]
```

```python
# 테스트 샘플을 사용하여 초기 이미지(에포크 0)를 생성 및 저장합니다.
generate_and_save_images(model, 0, test_sample)

for epoch in range(1, epochs + 1):

  start_time = time.time()

  for train_x in train_dataset:
    # 모델을 한 단계 학습
    train_step(model, train_x, optimizer)

  # 현재 에포크의 종료 시간을 기록
  end_time = time.time()

  # 손실을 계산하기 위해 평균 메트릭을 초기화
  loss = tf.keras.metrics.Mean()

  # 테스트 데이터셋의 각 배치에 대해 모델의 손실을 계산하고 평균에 추가
  for test_x in test_dataset:
    loss(compute_loss(model, test_x))

  # ELBO(Evidence Lower BOund) 값을 계산
  elbo = -loss.result()

  # 이전 출력을 지우고 새로운 정보를 화면에 표시
  display.clear_output(wait=False)

  # 에포크, ELBO, 그리고 현재 에포크에 걸린 시간을 출력.
  print('Epoch: {}, Test set ELBO: {}, time elapse for current epoch: {}'
        .format(epoch, elbo, end_time - start_time))

  # 현재 에포크의 결과 이미지를 생성 및 저장
  generate_and_save_images(model, epoch, test_sample)
```

### 마지막 훈련 epoch에서 생성된 이미지 표시하기

```python
def display_image(epoch_no):
  return PIL.Image.open('image_at_epoch_{:04d}.png'.format(epoch_no))
```

```python
# 애니메이션 파일의 이름을 'cvae.gif'로 설정
anim_file = 'cvae.gif'

# 'cvae.gif' 파일에 이미지를 쓰기 위한 imageio writer를 설정
with imageio.get_writer(anim_file, mode='I') as writer:
  # 'image*.png' 패턴에 맞는 모든 파일 이름을 가져옵니다.
  filenames = glob.glob('image*.png')
  # 파일 이름을 오름차순으로 정렬
  filenames = sorted(filenames)
  # 각 파일에 대해
  for filename in filenames:
    # 이미지를 읽어옵니다.
    image = imageio.imread(filename)
    # 읽어온 이미지를 gif 애니메이션에 추가
    writer.append_data(image)
  # 마지막 이미지를 다시 읽어옵니다. (애니메이션의 마지막 프레임을 재강조하기 위해)
  image = imageio.imread(filename)
  # 마지막 이미지를 gif 애니메이션에 추가합니다.
  writer.append_data(image)
```

tensorflow_docs.vis.embed 모듈의 embed_file 함수는 이미지 파일을 HTML에 임베드하는 함수입니다. 이 함수는 이미지 파일의 URL 또는 경로를 인자로 받아 HTML에 임베드할 태그를 반환합니다.  
Jupyter Notebook은 웹 기반의 대화식 컴퓨팅 환경이므로, 셀의 출력부에 나타나는 모든 것은 기본적으로 HTML로 렌더링될 수 있습니다.   
따라서, embed.embed_file(anim_file)를 Jupyter Notebook의 셀에서 실행하면, 해당 셀의 아래 출력 영역에 anim_file에 해당하는 애니메이션 또는 이미지가 나타나게 됩니다.

```python
import tensorflow_docs.vis.embed as embed
embed.embed_file(anim_file)  # anim_file에 지정된 이미지 파일을 HTML에 임베드
```

## Decoder 를 이용한 새로운 이미지 생성

```python
def generate_new_images(model, num_samples):
    """
    모델의 디코더를 사용하여 새로운 이미지를 생성합니다.

    Parameters:
    - model: 학습된 CVAE 모델
    - num_samples: 생성할 이미지의 수

    Returns:
    - generated_images: 생성된 이미지들
    """

    # 잠재 공간에서 무작위 벡터를 샘플링합니다.
    random_latent_vectors = tf.random.normal(shape=[num_samples, model.latent_dim])

    # 디코더를 사용하여 이미지를 생성합니다.
    generated_images_logits = model.decode(random_latent_vectors)
    generated_images = tf.sigmoid(generated_images_logits)

    return generated_images

# 사용 예제:
num_samples = 5  # 생성할 이미지의 수
new_images = generate_new_images(model, num_samples)

# 생성된 이미지들을 시각화
for i, image in enumerate(new_images):
    plt.subplot(1, num_samples, i + 1)
    plt.imshow(image[:, :, 0], cmap='gray')
    plt.axis('off')
plt.show()
```
