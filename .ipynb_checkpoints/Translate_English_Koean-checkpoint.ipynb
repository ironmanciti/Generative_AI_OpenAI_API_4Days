```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
from googletrans import Translator
# Initialize the translator
translator = Translator()

# Load the original CSV file
df = pd.read_csv("data/fine_food_reviews_1k.csv", index_col=0)
df.head()
```

```python
def preprocess_text(text):
    if pd.isna(text):
        return np.nan
    text = text.strip()  # 공백 제거
    text = ' '.join(text.split())  # 중복 공백 제거
    text = BeautifulSoup(text, "html.parser").get_text()  # HTML 태그 제거
    text = re.sub(r'[^\w\s]', '', text)  # 특수문자 및 이모지 제거
    return text

def translate_column(series):
    translated_summaries = []

    for i, text in enumerate(series.values):
        try:
            preprocessed_text = preprocess_text(text)  # 전처리
            if pd.isna(preprocessed_text):
                translated_summaries.append(np.nan)
            else:
                translated_text = translator.translate(preprocessed_text, src='en', dest='ko').text
                translated_summaries.append(translated_text)
        except Exception as e:
            translated_summaries.append(np.nan)
            print(f"Translation failed for text: {text} with error: {e}")
        
        # 100건마다 진행 상황 출력
        if (i + 1) % 100 == 0:
            print(f"{i + 1}건 처리 완료")
    
    return translated_summaries
```

```python
translated_summaries = translate_column(df['Summary'])
```

```python
# Assign the translated summaries to the DataFrame
df['Summary'] = translated_summaries
df.head()
```

```python
translated_summaries = translate_column(df['Text'])
```

```python
df['Text'] = translated_summaries
df.head()
```

```python
df.dropna(inplace=True)
```

```python
# Save the translated DataFrame to a new CSV file
df.to_csv("data/fine_food_reviews_1k_fully_translated_korean.csv")
```
